```python
import json 
import sys
import numpy as np
```

```python
# LOADING RAW DATA
raw_data = open('../data/exp_1_one_player_known_structure/main/raw_data/main_1_no_hint_full.json')
raw_data = json.load(raw_data)
```

```python
## EXTRACTING ONLY RELEVANT INFOR FOR SINGLE PLAYER
def findkeys(node, kv):
    if isinstance(node, list):
        for i in node:
            for x in findkeys(i, kv):
                yield x
    elif isinstance(node, dict):
        if kv in node:
            yield node[kv]
        for j in node.values():
            for x in findkeys(j, kv):
                yield x

data = {}  
complete_data = {}
complete_data_no_debrief = {}
rooms = list(raw_data.keys())
room_numbers = {}

print('Number of rooms: ', len(rooms)) # number of triads / rooms 
for room in rooms[2:]:   # looping over rooms and each entry in a room (potentially skipping initial rooms if they were sanity checks not belonging to pp)
    data[room] = {}
    room_keys = list(raw_data[room].keys()) 
    demographic_data = list(findkeys(raw_data[room], 'debriefData'))
    room_number = list(findkeys(raw_data[room], 'roomNumber'))[0]
    room_numbers[room] = room_number
    for room_key in room_keys[:len(room_keys)]:     
        if 'planetSelections' in raw_data[room][room_key].keys():
            subj_ids = list(raw_data[room][room_key]['planetSelections'].keys())   
            room_data = {}
            subj_data = {}
            room_complete = 0
            count = 1
            for subj in subj_ids:
                if len(raw_data[room][room_key]['participants']) == 3:
                    if len(raw_data[room][room_key]['planetSelections'][subj]) >= 11: # 10 for conditions in which struct is known 
                        subj_data[subj] = []
                        room_data['subj_ids'] = []
                        room_data['roomNumber'] = room_number
                        room_data['subj_training'] = []
                        room_data['subj_social_training'] = []
                        room_data['subj_social_training_2'] = []
                        room_data['PID'] = []
                        room_data['subj_names'] = []
                        room_data['condition'] = []
                        room_data['fish'] = []
                        room_data['fish_baseline'] = [0, 1, 0]
                        room_data['subject_function'] = []
                        room_data['positions'] = []
                        for nested_subj in raw_data[room][room_key]['participants']:
                            if nested_subj['id'] in list(raw_data[room][room_key]['planetSelections'].keys()):
                                pidkey = "pid"
                                if pidkey not in list(raw_data[room][room_key]['planetSelections'][nested_subj['id']][0]['participantPID'].keys()):
                                    pidkey = "PID"
                                room_data['PID'].append(raw_data[room][room_key]['planetSelections'][nested_subj['id']][0]['participantPID'][pidkey])
                                room_data['subj_names'].append(nested_subj['name'])
                                room_data['subject_function'].append(nested_subj['stimuli'][0])
                                room_data['condition'].append(nested_subj['stimuli'][1])
                                room_data['fish'].append(nested_subj['stimuli'][2])
                                room_data['positions'].append(nested_subj['stimuli'][3]) 
                                room_data['subj_training'].append(raw_data[room][room_key]['planetSelections'][nested_subj['id']][0]['training']['training'])
                                room_data['subj_social_training'].append(raw_data[room][room_key]['planetSelections'][nested_subj['id']][0]['socialTraining']['socialTraining'])
                                room_data['subj_social_training_2'].append(raw_data[room][room_key]['planetSelections'][nested_subj['id']][0]['socialTrainingStructure']['socialTrainingStructure'])    
                            data[room]['room_info'] = room_data
                            data[room]['subj_responses'] = subj_data
                    if len(raw_data[room][room_key]['planetSelections'][subj]) <= 11:
                        subj_data[subj] = raw_data[room][room_key]['planetSelections'][subj][:len(raw_data[room][room_key]['planetSelections'][subj])]
                    # if len(raw_data[room][room_key]['planetSelections'][subj]) <= 10: # uncomment this if structure is known 
                        # subj_data[subj] = raw_data[room][room_key]['planetSelections'][subj][:len(raw_data[room][room_key]['planetSelections'][subj])]
        if 'room_info' in data[room].keys():
            complete_data[room] = {}
            complete_data[room]['room_info'] = data[room]['room_info']
            complete_data[room]['subj_responses'] = data[room]['subj_responses']
            if demographic_data != []:
                complete_data[room]['demographics'] = demographic_data[0]

# uncomment the below and apply some changes if we are dealing with mulitple players 
#     room_complete_check = 3
#     demographics = {} 
#     if room_complete >= room_complete_check:
#         room_complete_2 = []
#         for room_key in room_keys: # demographics           
#             if 'debriefData' in list(raw_data[room][room_key].keys()):
#                 room_complete_2.append(1)
#                 subj_demographic = raw_data[room][room_key]['debriefData']
#                 demographics[raw_data[room][room_key]['participantId']] = subj_demographic  
#             if sum(room_complete_2) >= room_complete_check:
#                 complete_data[room] = {}
#                 complete_data[room]['room_info'] = room_data
#                 complete_data[room]['subj_responses'] = subj_data
#                 complete_data[room]['demographics'] = demographics 
#             complete_data_no_debrief[room] = {}
#             complete_data_no_debrief[room] = {}
#             complete_data_no_debrief[room]['room_info'] = room_data
#             complete_data_no_debrief[room]['subj_responses'] = subj_data       
#             complete_data_no_debrief[room]['demographics'] = demographics 
print('number of complete rooms', len(complete_data))
```

```python
# TRANSFORMING DATA TO CLEAN FORMAT 
clean_data = {}
print(complete_data.keys())
for room in list(complete_data.keys()):
    clean_data[room] = {}
    subjs_planet_selections = {}
    subjs_structure_selections = {}
    if 'fish' in list(complete_data[room]['room_info'].keys()):
        subj_fish = complete_data[room]['room_info']['fish']
    if len(subj_fish) == 3:
        participant_keys = {'subject1': 0, 'subject2': 1, 'subject3': 2}
        for subj in complete_data[room]['subj_responses']:
            subj_responses = []
            subj_sim_responses = []
            subj_info = []
            for response in complete_data[room]['subj_responses'][subj]:
                if 'simulatedResponse' in list(response.keys()):
                    sim_resp = response['simulatedResponse']
                    color = response['color']
                    confidence = response['confidence']
                    if 'demographics' in complete_data[room].keys():
                        trust = complete_data[room]['demographics']['trust']
                        age = complete_data[room]['demographics']['age']
                        gender = complete_data[room]['demographics']['gender']
                        comment =  complete_data[room]['demographics']['comment']
                    else:
                        trust, age, gender, comment = 1,1,1,1
                    pp_number = response['participantNumber']
                    if 'globalCondition' in response.keys():
                        condition = response['globalCondition']
                    condition = list(findkeys(complete_data[room], 'globalCondition'))[0]
                    structureHint = list(findkeys(complete_data[room], 'structureHint'))[0]
                    fish = subj_fish[participant_keys[pp_number]]
                    if fish == 1:
                        fish = 'blue'  
                    elif fish == 0:               
                        fish = 'red'
                    if 'PID' in response['participantPID'].keys():
                        pid = response['participantPID']['PID']
                    else:
                        pid = response['participantPID']['pid']
                    training = response['training']['training']
                    social_training = response['socialTraining']['socialTraining']
                    social_training_2 = response['socialTrainingStructure']['socialTrainingStructure']
                    if pp_number == 'subject1':
                        t_fish = 1
                    elif pp_number == 'subject2':
                        t_fish = 3
                    elif pp_number == 'subject3':
                        t_fish = 6
                    # flip confidence rating if color was red as currently red is on the left and we start with -3
                    if color == 'deepskyblue':
                        if confidence == 1: 
                            confidence = -1
                        elif confidence == 2:
                            confidence = -2
                        elif confidence == 3:
                            confidence = -3 
                    # now need to flip selections when initial fish for pp was blue 
                    if fish == 'blue' and pp_number != 'subject2':
                        confidence *= -1
                    if fish == 'red' and pp_number == 'subject2':
                        confidence *= -1
                    subj_responses.append(confidence)
                    subj_sim_responses.append(sim_resp)
                    if subj_info == []:       
                        structure_selection = np.zeros((3,3))
                        if condition == 'independent':
                            ground_truth_structure = [[0, 0, 0], [1, 0, 0], [1, 0, 0]]
                        elif condition == 'lr':
                            ground_truth_structure = [[0, 0, 0], [1, 0, 1], [1, 0, 0]]
                        if pp_number == 'subject1':
                            if len(complete_data[room]['subj_responses'][subj]) == 11:
                                structure_selection[0,1] = complete_data[room]['subj_responses'][subj][10]['causalStructure']['youTwoDown']['youTwoDown']
                                structure_selection[0,2] = complete_data[room]['subj_responses'][subj][10]['causalStructure']['youThreeDown']['youThreeDown']
                                structure_selection[1,0] = 1
                                structure_selection[2,0] = 1
                                structure_selection[1,2] = complete_data[room]['subj_responses'][subj][10]['causalStructure']['twoThreeRight']['twoThreeRight']
                                structure_selection[2,1] = complete_data[room]['subj_responses'][subj][10]['causalStructure']['threeTwoLeft']['threeTwoLeft']
#                         elif pp_number == 'subject2': # uncomment this for multiple players
#                             if len(complete_data[room]['subj_responses'][subj]) == 11:
#                                 if room != 'ZnNeDf90UL' and room != 'I5LM0q70aN' and room != 'evFhLqXraB':
#                                     structure_selection[0,1] = 0
#                                     structure_selection[0,2] = complete_data[room]['subj_responses'][subj][10]['causalStructure']['twoThreeRight']['twoThreeRight']
#                                     structure_selection[1,0] = complete_data[room]['subj_responses'][subj][10]['causalStructure']['youTwoDown']['youTwoDown']
#                                     structure_selection[2,0] = complete_data[room]['subj_responses'][subj][10]['causalStructure']['threeTwoLeft']['threeTwoLeft']
#                                     structure_selection[1,2] = complete_data[room]['subj_responses'][subj][10]['causalStructure']['youThreeDown']['youThreeDown']
#                                     structure_selection[2,1] = 0
#                         elif pp_number == 'subject3':
#                             if len(complete_data[room]['subj_responses'][subj]) == 11:
#                                 if condition == 'independent':
#                                     structure_selection[0,1] = complete_data[room]['subj_responses'][subj][10]['causalStructure']['twoThreeRight']['twoThreeRight']
#                                     structure_selection[0,2] = 0
#                                     structure_selection[1,0] = complete_data[room]['subj_responses'][subj][10]['causalStructure']['threeTwoLeft']['threeTwoLeft']
#                                     structure_selection[2,0] = complete_data[room]['subj_responses'][subj][10]['causalStructure']['youTwoDown']['youTwoDown']
#                                     structure_selection[1,2] = 0
#                                     structure_selection[2,1] = complete_data[room]['subj_responses'][subj][10]['causalStructure']['youThreeDown']['youThreeDown']     
#                                 elif condition == 'lr':
#                                     structure_selection[0,1] = complete_data[room]['subj_responses'][subj][10]['causalStructure']['twoThreeRight']['twoThreeRight']
#                                     structure_selection[0,2] = 0
#                                     structure_selection[1,0] = complete_data[room]['subj_responses'][subj][10]['causalStructure']['threeTwoLeft']['threeTwoLeft']
#                                     structure_selection[2,0] = complete_data[room]['subj_responses'][subj][10]['causalStructure']['youTwoDown']['youTwoDown']
#                                     structure_selection[1,2] = 1
#                                     structure_selection[2,1] = complete_data[room]['subj_responses'][subj][10]['causalStructure']['youThreeDown']['youThreeDown']            
                        subj_info = {'number': pp_number, 'structureHint': structureHint, 'condition': condition, 'fish': fish, 't_fish': t_fish, 'pid': pid, 'selected_structure': structure_selection.tolist(), 'training': training, 'socialTraining': social_training, 'socialTrainingStructure': social_training_2, 'trust': trust, 'age': age, 'gender': gender, 'comment': comment, 'ground_truth_structure': ground_truth_structure}
            subjs_planet_selections[subj] = {'room_number': room_numbers[room] ,'info': subj_info, 'planet_selections': subj_responses, 'sim_resp': subj_sim_responses}
            if len(complete_data[room]['subj_responses'][subj]) == 11:
                subjs_structure_selections[subj] = complete_data[room]['subj_responses'][subj][10]
        clean_data[room] = subjs_planet_selections 
# with open("../data/exp_1_one_player_known_structure/main/NAMEHERE", "w") as datafile: 
#     json.dump(clean_data, datafile)
```
